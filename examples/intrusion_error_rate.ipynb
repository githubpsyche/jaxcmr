# Intrusion error rate

Compute intrusion error rates from recall data.

This notebook demonstrates how to compute the intrusion error rate, the proportion of recalled items that were not part of the study list. The example uses a toy dataset to illustrate the metric and its visualization. Such rates help assess memory accuracy by highlighting spurious recalls.

```python
import jax.numpy as jnp
from jaxcmr.analyses.intrusion_error_rate import (
    trial_intrusion_error_rate,
    intrusion_error_rate,
    plot_intrusion_error_rate,
)
import matplotlib.pyplot as plt
```

```python
recalls = jnp.array([[4, 1, 2], [1, 2, 3]], dtype=jnp.int32)
presentations = jnp.array([[1, 2, 3], [1, 2, 3]], dtype=jnp.int32)

dataset = {
    'recalls': recalls,
    'pres_itemnos': presentations,
    'listLength': jnp.array([[3], [3]], dtype=jnp.int32),
    'subject': jnp.array([[1], [2]], dtype=jnp.int32),
}
trial_mask = jnp.array([True, True])
```

```python
intrusion_error_rate(recalls, presentations, list_length=3)
```

```python
plot_intrusion_error_rate(dataset, trial_mask)
plt.show()
```
