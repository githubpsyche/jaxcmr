# Order Error Rate


Demonstrate computing order error rates.


Order errors occur when a list item is recalled in the wrong output position. This notebook shows how to quantify those errors in serial-recall data using small toy arrays.


```python
import jax.numpy as jnp
from jaxcmr.analyses.order_error_rate import order_error_rate, trial_order_error_rate
```

```python
recalls = jnp.array([[2, 1, 3], [1, 2, 3]], dtype=jnp.int32)
presentations = jnp.array([[1, 2, 3], [1, 2, 3]], dtype=jnp.int32)
list_length = 3
```

```python
trial_errors = trial_order_error_rate(recalls[0], presentations[0])
error_rates = order_error_rate(recalls, presentations, list_length)
error_rates, trial_errors
```
