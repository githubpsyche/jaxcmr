# Category SPC

> Compute and visualize category-filtered serial position curves.

The category SPC restricts the serial position curve to items of a specific category, revealing category-dependent encoding and retrieval patterns across study positions.

This notebook is parameterized for use with [papermill](https://papermill.readthedocs.io/).

| Parameter | Description |
|-----------|-------------|
| `data_path` | Path to HDF5 dataset file |
| `figure_dir` | Directory for saving figures |
| `figure_str` | Filename for saved figure (empty string to skip saving) |
| `ylim` | Y-axis limits as tuple, or `None` for auto |
| `trial_query` | Query string for trial selection |
| `category_field` | Key providing item categories per study position |
| `contrast_name` | Legend title for contrasts |
| `category_values` | Category values to plot |
| `labels` | Legend labels per category |
| `color_cycle` | Colors for each curve |

```python
#| include: false
from nbdev.showdoc import show_doc
```

```python
#| code-summary: Imports and setup
import os
import warnings

import jax.numpy as jnp
import matplotlib.pyplot as plt

from jaxcmr.analyses.cat_spc import cat_spc, plot_cat_spc, category_recall_counts
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data

warnings.filterwarnings("ignore")


def save_figure(figure_dir, figure_str, suffix=None):
    plt.tight_layout()
    if not figure_str:
        plt.show()
        return
    os.makedirs(figure_dir, exist_ok=True)
    suffix_str = f"_{suffix}" if suffix else ""
    figure_path = os.path.join(figure_dir, f"{figure_str}{suffix_str}.png")
    plt.savefig(figure_path, bbox_inches="tight", dpi=600)
    plt.show()
```

```python
#| code-summary: Parameters for this analysis

run_tag = "Category SPC"
data_path = "data/TalmiEEG.h5"
figure_dir = "results/figures"
figure_str = ""
ylim = None
trial_query = "data['subject'] > 0"
category_field = "condition"
category_values = [1, 2]
labels = ["Negative", "Neutral"]
color_cycle = ["red", "black"]
contrast_name = "condition"
```

```python
#| code-summary: Load data and generate trial mask
project_root = find_project_root()
figure_dir = os.path.join(project_root, figure_dir)
data_path = os.path.join(project_root, data_path)
data = load_data(data_path)
trial_mask = generate_trial_mask(data, trial_query)
```

```python
#| code-summary: Plot category-filtered serial position curve
#| label: fig-cat-spc
#| fig-cap: "Category-filtered serial position curve."
plot_cat_spc(data, trial_mask, category_field=category_field, category_values=category_values, labels=labels, color_cycle=color_cycle, contrast_name=contrast_name)
if ylim is not None:
    for ax in plt.gcf().axes:
        ax.set_ylim(ylim)
save_figure(figure_dir, figure_str)
```
