# Guide

Core abstractions and patterns for jaxcmr

jaxcmr models are JAX PyTrees—nested structures of arrays that integrate with JAX transformations like `jit` and `vmap`. Model methods return new instances rather than mutating state, enabling reproducible simulations and efficient batching. See [Why JAX](jax.ipynb) for details on how this works.
