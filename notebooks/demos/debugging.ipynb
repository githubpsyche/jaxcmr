Using the `jax.disable_jit()` context seems to be a great way to make JAX code easier to debug, providing transparency into variable values and stack traces during normal execution.

```python
import jax

@jax.jit
def f(x):
    y = x * 2
    return y + 3

with jax.disable_jit():
    print(jax.lax.map(f, jax.numpy.array([1, 2, 3])))
```
