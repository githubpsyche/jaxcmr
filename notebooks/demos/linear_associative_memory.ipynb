```python
from jaxcmr.models import LinearAssociativeMemory
import jax.numpy as jnp
from jax import jit

item_count = 16
learning_rate = 0.1
items = jnp.eye(item_count, dtype=jnp.float32)
```

```python
test = lambda: LinearAssociativeMemory.initialize_mfc(item_count, learning_rate)

@jit
def test():
    return LinearAssociativeMemory.initialize_mfc(item_count, learning_rate)

%timeit test()
```

```python
test = lambda : LinearAssociativeMemory.initialize_mfc(items, learning_rate)

@jit
def test():
    return LinearAssociativeMemory.initialize_mfc(items, learning_rate)

%timeit test()
```

```python
import jax

with jax.disable_jit():
    @jax.jit
    def basic_init_mfc(
        item_count,
        learning_rate
        ) -> LinearAssociativeMemory.LinearAssociativeMfc:
        "A linear associative feature-to-context memory assuming one-hot item representations"
        memory = jnp.eye(item_count, item_count + 2)
        memory = jnp.hstack([jnp.zeros((item_count, 1)), memory[:, :-1]])
        memory = memory * (1 - learning_rate)
        return LinearAssociativeMemory.LinearAssociativeMfc(memory)

    basic_init_mfc(item_count, learning_rate)

basic_init_mfc(item_count, learning_rate)
```
