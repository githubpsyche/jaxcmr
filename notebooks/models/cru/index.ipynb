# CRU Models

Context Retrieval and Updating for serial recall


CRU (Context Retrieval and Updating) is a streamlined retrieved-context model developed by Gordon Logan for automatic sequential behaviors like typing and serial recall. jaxcmr implements CRU variants alongside CMR to explore their relationship.

## CRU vs CMR: A Spectrum

CRU and CMR share the core idea of drifting temporal context, but differ in complexity:

| Aspect | CRU | CMR |
|--------|-----|-----|
| **Design goal** | Serial recall, typing | Free recall |
| **Architecture** | Streamlined | Full-featured |
| **Feature-to-context (MFC)** | Omitted ($\gamma=0$) | Dynamic learning |
| **Pre-experimental MCF** | Omitted ($\delta=\alpha=0$) | Explicit support |
| **Context input** | Direct item → context | Retrieved via MFC |
| **Recall termination** | Context-based (end-of-list) | Position-based |
| **Item identification** | Confusability mechanism | Perfect recognition |

## Why CRU?

CRU was developed to explain how sequential behavior becomes automatic:

1. **Typing**: Producing letter sequences from memory
2. **Serial recall**: Reproducing lists in order
3. **Hierarchical control**: Top-level goals generating action sequences

For these strictly ordered tasks, CRU's simpler architecture suffices—and provides insights into what CMR mechanisms are truly necessary.

## The Factorial Approach

jaxcmr implements CRU and CMR as points on a spectrum, allowing factorial comparison:

```python
Pure CRU ←——— Hybrid Variants ———→ Pure CMR
   │                                    │
   └── Add CMR features one by one ────┘
```

By selectively enabling CMR mechanisms in CRU, we can identify which features matter for which tasks.

## Key CRU Mechanisms

### No Feature-to-Context Learning

CRU treats the MFC as fixed (pre-experimental only):

$$\mathbf{c}^{IN}_i = \mathbf{f}_i$$

The item representation directly becomes the context input, with no experimental associations.

**Consequence**: No backward transitions from context reinstatement. CRU predicts strongly forward-going recall.

### No Pre-Experimental MCF Support

CRU omits the `shared_support` ($\alpha$) and `item_support` ($\delta$) parameters:

$$M^{CF}_{pre} = 0$$

All context-to-item associations come from experimental learning.

### Item Identification Mechanism

CRU includes a probabilistic item identification stage absent from CMR:

$$v_i = \exp[-g \cdot d_{ij}]$$

where:
- $v_i$ = drift rate for identifying item $i$
- $g$ = sensitivity parameter
- $d_{ij}$ = distance between presented item and candidate $j$

This handles confusable stimuli (e.g., similar letters) and explains intrusion errors in serial recall.

### Serial Position Scaling

CRU modulates parameters by serial position:

$$g_i = g_{max} \cdot g_{dec}^{i-1}$$
$$\beta_i = \beta_{max} \cdot \beta_{dec}^{i-1}$$

This creates primacy effects through encoding dynamics rather than MCF learning rates.

### Context-Based Termination

CRU treats end-of-list as a special item associated with final context:

```python
Study: [A, B, C, END]
       ↓  ↓  ↓   ↓
       c₁ c₂ c₃  c₄
```

At retrieval, END competes with items. When context drifts toward the end-of-list state, termination becomes likely.

**Contrast with CMR**: CMR uses position-based stopping ($P_{stop} = \theta_s e^{j\theta_r}$).

## When To Use CRU

Use CRU when:
- Modeling serial recall tasks
- Working with confusable stimuli (letters, similar items)
- Testing whether CMR's complexity is necessary
- Studying automatic sequential behavior

Use CMR when:
- Modeling free recall
- Backward transitions matter
- Flexible recall initiation is needed
- Semantic organization is important

## Empirical Findings

Research comparing CRU and CMR (see [projects/cru_to_cmr](../../../projects/cru_to_cmr/)) shows:

**For Free Recall:**
- CRU needs CMR-like features to capture benchmark phenomena
- Dynamic MFC ($\gamma > 0$) is crucial for backward transitions
- Flexible recall initiation ($\beta_{start}$) captures recency

**For Serial Recall:**
- CRU's context-based termination outperforms CMR's position-based rule
- Item identification mechanism captures intrusion patterns
- CMR's added mechanisms provide modest improvements
