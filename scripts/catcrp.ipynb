```python
import os
import warnings
import numpy as np
from jaxcmr.analyses.distcrp import plot_dist_crp
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
from jaxcmr.math import cosine_similarity_matrix
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
run_tag = "SemCRP"
data_tag = "CohenKahana2022_noELI"
data_path = "data/CohenKahana2022_noELI.h5"
data_query = "data['session'] == 1"
embedding_path = "data/florida_nouns-all-mpnet-base-v2.npy"
```

```python
data = load_data(os.path.join(find_project_root(), data_path))
trial_mask = generate_trial_mask(data, data_query)

embeddings = np.load(os.path.join(find_project_root(), embedding_path)).astype(np.float32)
distances = 1 - cosine_similarity_matrix(embeddings)
```

```python
plot_dist_crp(data, trial_mask, distances=distances)
plt.show()
```
