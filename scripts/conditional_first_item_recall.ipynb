# First Item Recall Demo

A short example showing how to compute and plot the probability that the first studied item is recalled at each output position.

```python
import os
import warnings

from matplotlib import rcParams  # type: ignore
import matplotlib.pyplot as plt

from jaxcmr.analyses.nth_item_recall import plot_conditional_nth_item_recall_curve
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data

warnings.filterwarnings("ignore")
```

```python
run_tag = "FirstItemRecall"
data_name = "HealeyKahana2014"
data_query = "data['listtype'] == -1"
data_path = "data/HealeyKahana2014.h5"
```

```python
data = load_data(os.path.join(find_project_root(), data_path))
trial_mask = generate_trial_mask(data, data_query)
```

```python
color_cycle = [each["color"] for each in rcParams["axes.prop_cycle"]]
i = 0
maximum = 2

axis = plot_conditional_nth_item_recall_curve(
    data,
    trial_mask,
    query_study_position=i,
    labels=[str(i+1)],
    color_cycle=color_cycle,
    contrast_name="Position",
)
# plt.show()

for i in range(i+1, maximum):
    axis = plot_conditional_nth_item_recall_curve(
        data,
        trial_mask,
        query_study_position=i,
        labels=[str(i+1)],
        color_cycle=color_cycle,
        contrast_name="Position",
        axis=axis
    )

plt.ylabel("P(Item | Available & Continuing)")
# plt.ylim([0, .5])
# plt.show()
```
