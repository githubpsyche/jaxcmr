# Categorized Serial Position Curve
Plot SPC curves factored by categorical features (e.g., emotion tags).

```python
params = {
    "data_name": "TalmiEEG",
    "data_query": "data['subject'] > 0",
    "category_field": "condition",
    "contrast_name": "condition",
    "category_values": [1, 2],
    "labels": ["Negative", "Neutral"],
    "color_cycle": ["red", "black"],
    "ylim": [0, 1],
    "output_path": "projects/TalmiEEG/results/figures/reference_catspc.png",
}

pm.execute_notebook(
    f"{find_project_root()}/templates/cat_spc.ipynb",
    f"{find_project_root()}/projects/TalmiEEG/cat_spc.ipynb",
    parameters=params,
)

print(f"![]({params["output_path"]})")
display(Image(filename=params["output_path"]))
```

# Categorized LPP by Study Position

Plot LPP amplitude by study position, factored by categorical features (e.g., emotion tags).

```python
params = {
    "data_name": "TalmiEEG",
    "data_query": "data['subject'] > 0",
    "category_field": "condition",
    "category_values": [1, 2],
    "labels": ["Negative", "Neutral"],
    "color_cycle": ["red", "black"],
    "ylim": [0, 1.5],
    "output_path": "projects/TalmiEEG/results/figures/cat_lpp_spc_{}.png",
}

for lpp_field in ["EarlyLPP", "LateLPP"]:
    pm.execute_notebook(
        f"{find_project_root()}/templates/cat_lpp_spc.ipynb",
        f"{find_project_root()}/projects/TalmiEEG/cat_lpp_spc_{lpp_field}.ipynb",
        parameters=params
        | {
            "lpp_field": lpp_field,
            "output_path": params["output_path"].format(lpp_field),
        },
    )
    print(f"![]({params["output_path"].format(lpp_field)})")
    display(Image(filename=params["output_path"].format(lpp_field)))
```

# Categorized Recall Rate by LPP

Plot the recall rate as a function of binned LPP amplitude for different categories.


```python
params = {
    "data_name": "TalmiEEG",
    "data_query": "data['subject'] > 0",
    "category_field": "condition",
    "category_values": [1, 2],
    "labels": ["Negative", "Neutral"],
    "color_cycle": ["red", "black"],
    "ylim": [0, .8],
    "output_path": "projects/TalmiEEG/results/figures/cat_recall_by_lpp_{}.png",
}

for lpp_field in ["EarlyLPP", "LateLPP"]:
    pm.execute_notebook(
        f"{find_project_root()}/templates/cat_recall_by_lpp.ipynb",
        f"{find_project_root()}/projects/TalmiEEG/cat_recall_by_lpp_{lpp_field}.ipynb",
        parameters=params
        | {
            "lpp_field": lpp_field,
            "output_path": params["output_path"].format(lpp_field),
        },
    )
    print(f"![]({params["output_path"].format(lpp_field)})")
    display(Image(filename=params["output_path"].format(lpp_field)))

    print(f"![]({params["output_path"].replace('.png', '_counts.png').format(lpp_field)})")
    display(Image(filename=params["output_path"].replace('.png', '_counts.png').format(lpp_field)))
```
