```python
import papermill as pm
from jaxcmr.helpers import find_project_root
from IPython.display import Image, display
```

# Categorized Serial Position Curve
Plot SPC curves factored by categorical features (e.g., emotion tags).

```python
data_name = "TalmiEEG"
data_query =  "data['subject'] > 0"
category_field = "condition"
contrast_name = "condition"
category_values = [1, 2]
labels = ["Negative", "Neutral"]
color_cycle = ["red", "black"]
ylim = [0, 1]
output_path = "projects/TalmiEEG/results/figures/reference_catspc.png"

pm.execute_notebook(
    f"{find_project_root()}/templates/cat_spc.ipynb",
    f"{find_project_root()}/projects/TalmiEEG/cat_spc.ipynb",
    parameters=dict(
        data_name=data_name,
        data_query=data_query,
        category_field=category_field,
        contrast_name=contrast_name,
        category_values=category_values,
        labels=labels,
        color_cycle=color_cycle,
        ylim=ylim,
        output_path=output_path,
    ),
)

print(output_path)
display(Image(filename=output_path))
```

# Categorized LPP by Study Position

Plot LPP amplitude by study position, factored by categorical features (e.g., emotion tags).

```python
data_name = "TalmiEEG"
data_query =  "data['subject'] > 0"
category_field = "condition"
category_values = [1, 2]
labels = ["Negative", "Neutral"]
color_cycle = ["red", "black"]
ylim = [0, 1.5]
output_path = "projects/TalmiEEG/results/figures/cat_splpp_{}.png"

for lpp_field in ["EarlyLPP", "LateLPP"]:
    pm.execute_notebook(
        f"{find_project_root()}/templates/cat_splpp.ipynb",
        f"{find_project_root()}/projects/TalmiEEG/cat_splpp_{lpp_field}.ipynb",
        parameters=dict(
            data_name=data_name,
            data_query=data_query,
            category_field=category_field,
            category_values=category_values,
            lpp_field=lpp_field,
            labels=labels,
            color_cycle=color_cycle,
            ylim=ylim,
            output_path=output_path.format(lpp_field.lower()),
        ),
    )
    print(output_path.format(lpp_field.lower()))
    display(Image(filename=output_path.format(lpp_field.lower())))
```
