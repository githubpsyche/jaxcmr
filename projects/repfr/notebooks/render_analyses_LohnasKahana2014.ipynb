```python
import os
import numpy as np
import papermill as pm

from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
```

# RepFR Analyses Runner
Run minimal analyses via papermill and save rendered notebooks.


```python
project_root = find_project_root()
templates_dir = os.path.join(project_root, "templates")
rendered_dir = os.path.join(project_root, "projects/repfr/notebooks/rendered")
figures_dir = os.path.join(project_root, "projects/repfr/results/figures/analyses")

os.makedirs(rendered_dir, exist_ok=True)
os.makedirs(figures_dir, exist_ok=True)

data_name = "LohnasKahana2014"
data_path = os.path.join(project_root, "data", f"{data_name}.h5")
data = load_data(data_path)

trial_query = "data['list_type'] > 0"
control_query = "data['list_type'] == 1"
repcrp_query = "data['list_type'] == 4"
```

## Analyses


```python
def list_type_tag(query: str) -> str:
    trial_mask = generate_trial_mask(data, query)
    list_types = np.unique(np.array(data["list_type"][trial_mask]).astype(int))
    types_str = "".join(str(t) for t in list_types)
    return f"list_type_{types_str}" if types_str else "list_type_all"

analysis_specs = [
    {
        "name": "spc",
        "template": "spc.ipynb",
        "rendered": "spc.ipynb",
        "parameters": {
            "run_tag": "SPC",
            "data_name": data_name,
            "data_query": trial_query,
            "output_path": os.path.join(figures_dir, "spc.png"),
        },
    },
    {
        "name": "crp",
        "template": "crp.ipynb",
        "rendered": "crp.ipynb",
        "parameters": {
            "run_tag": "CRP",
            "data_name": data_name,
            "data_query": trial_query,
            "output_path": os.path.join(figures_dir, "crp.png"),
        },
    },
    {
        "name": "pnr",
        "template": "pnr.ipynb",
        "rendered": "pnr.ipynb",
        "parameters": {
            "run_tag": "PNR",
            "data_name": data_name,
            "data_query": trial_query,
            "query_recall_position": 0,
            "output_path": os.path.join(figures_dir, "pnr.png"),
        },
    },
    {
        "name": "repcrp",
        "template": "repcrp.ipynb",
        "rendered": "repcrp.ipynb",
        "parameters": {
            "data_name": data_name,
            "data_query": repcrp_query,
            "ctrl_query": control_query,
            "min_lag": 4,
            "output_path": os.path.join(figures_dir, "repcrp.png"),
        },
    },
]

for spec in analysis_specs:
    template_path = os.path.join(templates_dir, spec["template"])
    spec_data_name = spec["parameters"].get("data_name", data_name)
    query = spec["parameters"].get("data_query")
    tag = list_type_tag(query) if query else "list_type_all"
    rendered_name = f"{spec['name']}_{spec_data_name}_{tag}.ipynb"
    output_path = os.path.join(rendered_dir, rendered_name)
    print(f"Running {spec['name']} -> {output_path}")
    pm.execute_notebook(
        template_path,
        output_path,
        parameters=spec["parameters"],
    )
```
