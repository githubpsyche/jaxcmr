# Behavioral Analyses

Standard analyses for free recall data


jaxcmr provides functions for computing standard behavioral measures from free recall experiments. Each analysis can be applied to both empirical data and model simulations.

## Getting Started

- [Importing Data](importing-data.ipynb) - Loading recall datasets
- [Scoring Data](scoring-data.ipynb) - Computing basic metrics
- [Custom Analyses](custom-analyses.ipynb) - Implementing your own

## Analysis Categories

### Serial Position

How recall probability varies with study position.

| Analysis | Description |
|----------|-------------|
| [SPC](spc.ipynb) | Serial Position Curve - recall probability by position |
| [PNR](pnr.ipynb) | Probability of Nth Recall - what position is recalled first, second, etc. |

### Temporal Contiguity

How recall transitions relate to study order.

| Analysis | Description |
|----------|-------------|
| [CRP](crp.ipynb) | Conditional Response Probability - transition probabilities by lag |

### Category Effects

Effects of item categories on recall.

| Analysis | Description |
|----------|-------------|
| [CatCRP](catcrp.ipynb) | Category CRP - within vs. between category transitions |

### Repetition Effects

Analyses for experiments with repeated items.

| Analysis | Description |
|----------|-------------|
| [RepCRP](repcrp.ipynb) | Repetition CRP - transitions involving repeated items |

### Error Analyses

Patterns of recall errors.

| Analysis | Description |
|----------|-------------|
| [Intrusions](intrusion_error_rate.ipynb) | Extra-list intrusion rates |
| [Omissions](omission_error_rate.ipynb) | Items not recalled |
