```python
from jaxcmr.order_free_likelihood import enumerate_trial_permutations
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data

import itertools
import os
import numpy as np
```

```python
data_name = "HealeyKahana2014"
data_query = "np.logical_and(data['listtype'] == -1, data['subject'] == 63)"

```

```python
data_path = os.path.join(find_project_root(), f"data/{data_name}.h5")
data = load_data(data_path)
trial_mask = generate_trial_mask(data, data_query)

recalls = np.array(data["recalls"][trial_mask])
recalls.shape
```

```python
enumerate_trial_permutations(recalls)
```

```python
def pad_to_dense(M, maxlen: int) -> np.ndarray:
    """Appends the minimal required amount of zeroes at the end of each 
     array in the jagged array `M`, such that `M` looses its jagedness."""

    Z = np.zeros((len(M), maxlen))
    for enu, row in enumerate(M):
        Z[enu, :len(row)] += row 
    return Z
```

```python
%%timeit

nonzero_recalls = [
    trial_recalls[trial_recalls > 0].tolist() for trial_recalls in recalls
]

for nonzero, trial_recalls in zip(nonzero_recalls, recalls):
    unique_orders = itertools.permutations(nonzero)

pad_to_dense(list(unique_orders), trial_recalls.shape[0])
```
