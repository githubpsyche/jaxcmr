# Category-Based CRP

Plot distance-binned CRP curves using categorical features (e.g., emotion tags).


```python
import os

from jaxcmr.analyses.cat_spc import plot_cat_spc
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
```

```python
data_name = "TalmiEEG"
data_query =  "data['subject'] > 0"
category_field = "condition"
category_value = [1, 2]
labels = ["Negative", "Neutral"]
```

```python
project_root = find_project_root()
data_path = os.path.join(project_root, "data", f"{data_name}.h5")
data = load_data(data_path)
trial_mask = generate_trial_mask(data, data_query)
```

```python
axis = plot_cat_spc(
    datasets=[data, data],
    trial_masks=trial_mask,
    category_field=category_field,
    category_value=[1, 2],
    contrast_name=category_field,
    labels=["Negative", "Neutral"],
)
```
