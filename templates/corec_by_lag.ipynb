# Co-Recall by Lag


Quantify order-agnostic co-recall probability across study lags.


We treat each trial's recalls as a set of study positions, count how often pairs separated by lag d are recalled together, and average across trials. 
This template loads a dataset, selects trials with a simple query, and plots the resulting co-recall curve. 
Adjust the parameters below to examine other contrasts or lag limits.


```python
import os
import matplotlib.pyplot as plt
from jaxcmr.analyses.corec_by_lag import plot_corec_by_lag
from jaxcmr.helpers import find_project_root, generate_trial_mask, load_data
```

```python
run_tag = "CoRecByLag"
data_name = "LohnasKahana2014"
data_query = "data['list_type'] == 1"
max_lag = 5
ylim = [.1, .26]

data_path = os.path.join(find_project_root(), f"data/{data_name}.h5")
data = load_data(data_path)
trial_mask = generate_trial_mask(data, data_query)

plot_corec_by_lag(
    data,
    trial_mask,
    max_lag=max_lag,
)
plt.ylim(ylim)
```
